## Declination
How does the declination affect the uv-coverage.  
For convinience we use **SIMMS**

```python
from simms import simms # import simms
```

I want to simulate a 4 hour MeerKAT observation with a 30s integration at 1.4GHz using a 10MHz channel width. The pointing centre will vary (keep RA constant, only vary DEC)

```python
# Observation settings
observatory = "meerkat"
antennas = "MeerKAT64_ANTENNAS"
synthesis = 4
integration = 30
freq0 = "1400MHz"
dfreq = "10MHz"
nchan = 1
init_HA = -2
decs = [0,-20,-45,-90]
directions = ["J2000,0deg,%ddeg"%dec for dec in decs]
```

```python
for i,direction in enumerate(directions):
    label = 'meerkat_%s%d'%('m' if decs[i]<0 else 'p',abs(decs[i]))
    simms(tel=observatory,label=label,synthesis=synthesis,dtime=integration,start_time=init_HA,
          freq0=freq0,dfreq=dfreq,nchan=nchan,pos=antennas,direction=direction)
```

```python
def plot_uvcov(msname,select=None,figsize=[10,8],**kw):
    from pyrap.tables import table
    tab = table(msname)
    if select:
        # apply selection
        tab = tab.query(select)
    uvw = tab.getcol('UVW')
    plt.figure(1,figsize=figsize)
    plt.plot(-uvw[:,0],-uvw[:,1],'r.',**kw)
    plt.plot(uvw[:,0],uvw[:,1],'b.',**kw)
```

```python
for dec in decs:
    label = 'meerkat_%s%d'%('m' if dec<0 else 'p',abs(dec))
    plot_uvcov("%s_4h30.MS"%label,ms=.1)
    plt.show()
```

## Bonus Section
Instead of making multiple MSs, we can just make one MS with multiple pointings. 

```python
simms(tel=observatory,label='meerkat_multi',synthesis=synthesis,dtime=integration,start_time=init_HA,
          freq0=freq0,dfreq=dfreq,nchan=nchan,pos=antennas,direction=directions)    
```

```python
for field_id in range(len(directions)):
    plot_uvcov("meerkat_multi_4h30.MS",select='FIELD_ID==%d'%field_id,ms=.1)
    plt.show()
```
