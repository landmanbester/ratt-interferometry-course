# $uv$-coverage, sky models and visibilities

In this notebook we will be investigating:
* How to generate the $uv$-coverage of an EW-baseline.
* How to generate the $lm$-coordinates of a sky model.
* How to generate the visibilities that are produced by a skymodel and sample this visibility plane with the $uv$-track of our
EW baseline.

## $uv$-Coverage of an EW-baseline

When we have and EW baseline things simplify. Firstly: $XYZ = [0~d~0]^T$, where $d$ is the baseline length measured in wavelengths.
Moreover, we have that:
* $u = d\cos(H)$,
* $v = d\sin(H)\sin(\delta)$,

where $H$ is the hour angle of the field center and $\delta$ its declination. In this section we will be plotting the $uv$-coverage of an EW-baseline whose field center is at two different declinations.

```python
import numpy as np
import pylab as plt
%matplotlib inline
```

### Initialization 90$^{\circ}$

```python
H = np.linspace(-6,6,600)*(np.pi/12) #Hour angle in radians
d = 100 #We assume that we have already divided by wavelength
delta = 90*(np.pi/180) #Declination in radians
```

We have now set up the hour angle and the declination of our observation. Note that only the hour angle changes as the earth rotates. The field center and the NCP coincides.

```python
u_90 = d*np.cos(H)
v_90 = d*np.sin(H)*np.sin(delta)
```

Deriving the $uv$-track of our EW baseline over the observation period [-6h,6h].

```python
u_90_1 = d*np.cos(-6*(np.pi/12))
u_90_2 = d*np.cos(0*(np.pi/12))
u_90_3 = d*np.cos(6*(np.pi/12))

v_90_1 = d*np.sin(-6*(np.pi/12))*np.sin(delta)
v_90_2 = d*np.sin(0*(np.pi/12))*np.sin(delta)
v_90_3 = d*np.sin(6*(np.pi/12))*np.sin(delta)
```

Deriving specific points on the $uv$-track, the points where $H = \{$-6h,0h,6h$\}$

```python
plt.plot(u_90,v_90)
plt.hold("on")
plt.plot(u_90_1,v_90_1,"bo")
plt.plot(u_90_2,v_90_2,"go")
plt.plot(u_90_3,v_90_3,"ro")
plt.xlim([-d-10,d+10])
plt.ylim([-d-10,d+10])
```

Plotting the $uv$-track (over the 12 hour observation period). The three colored dots represent the three points that we calculated in the previous cell. The blue
dot can be associated with hour angle -6h, the green dot can be associated with hour angle 0, while the red dot can be associated with hour angle 6h. 
The $uv$-track is therefore moving in the anti-clockwise direction. Also note that the $uv$-track is a cricle with radius equal to $d$.

### Initialization 60$^{\circ}$

```python
delta = 60*(np.pi/180) #Declination in degrees
```

```python
u_60 = d*np.cos(H)
v_60 = d*np.sin(H)*np.sin(delta)
```

Deriving the $uv$-track of our EW baseline over the observation period [-6h,6h].

```python
u_60_1 = d*np.cos(-6*(np.pi/12))
u_60_2 = d*np.cos(0*(np.pi/12))
u_60_3 = d*np.cos(6*(np.pi/12))

v_60_1 = d*np.sin(-6*(np.pi/12))*np.sin(delta)
v_60_2 = d*np.sin(0*(np.pi/12))*np.sin(delta)
v_60_3 = d*np.sin(6*(np.pi/12))*np.sin(delta)
```

Deriving specific points on the $uv$-track, the points where $H = \{$-6h,0h,6h$\}$

```python
#delta 90
plt.plot(u_90,v_90)
plt.hold("on")
plt.plot(u_90_1,v_90_1,"bo")
plt.plot(u_90_2,v_90_2,"go")
plt.plot(u_90_3,v_90_3,"ro")
# delta 60
plt.plot(u_60,v_60,"r")
plt.plot(u_60_1,v_60_1,"bo")
plt.plot(u_60_2,v_60_2,"go")
plt.plot(u_60_3,v_60_3,"ro")
plt.xlim([-d-10,d+10])
plt.ylim([-d-10,d+10])
```

Plotting the $uv$-tracks (over the 12 hour observation period). The three colored dots represent $uv$-points at the three different hour angles. The blue dot can be associated with hour angle -6h, the green dot can be associated with hour angle 0, while the red dot can be associated with hour angle 6h. The $uv$-track is therefore moving in the anti-clockwise direction. The blue curve
corresponds to $\delta = 90^{\circ}$ and is a circle with radius equal to $d$. The red curve corresponds to $\delta = 60^{\circ}$ and is an ellipse with a semi-major axis equal to $d$ and a semi-minor axis equal to $d \sin(\delta)$.  

## Skymodel

Here we have a couple of sources given in RA ($\alpha$) and DEC ($\delta$):
* Source 1: (5h 32m 0.4s,60$^{\circ}$-17' 57'') - 1Jy
* Source 2: (5h 36m 12.8s,-61$^{\circ}$ 12' 6.9'') - 0.5Jy
* Source 3: (5h 40m 45.5s,-61$^{\circ}$ 56' 34'') - 0.2Jy

The field center is located at $(\alpha_0,\delta_0) = $ (5h 30m,60$^{\circ}$).  

```python
RA_sources = np.array([5+30.0/60,5+32.0/60+0.4/3600,5+36.0/60+12.8/3600,5+40.0/60+45.5/3600])
DEC_sources = np.array([60,60+17.0/60+57.0/3600,61+12.0/60+6.9/3600,61+56.0/60+34.0/3600])
Flux_sources_labels = np.array(["","1 Jy","0.5 Jy","0.2 Jy"])
Flux_sources = np.array([1,0.5,0.1]) #in Jy
step_size = 200
print RA_sources
print DEC_sources
```

We have now converted the celestial coordinates into degrees

```python
RA_rad = np.array(RA_sources)*(np.pi/12)
DEC_rad = np.array(DEC_sources)*(np.pi/180)
RA_delta_rad = RA_rad-RA_rad[0]

l = np.cos(DEC_rad)*np.sin(RA_delta_rad)
m = (np.sin(DEC_rad)*np.cos(DEC_rad[0])-np.cos(DEC_rad)*np.sin(DEC_rad[0])*np.cos(RA_delta_rad))
print l*(180/np.pi)
print m*(180/np.pi)
```

Recall that we can convert RA/DEC to $lm$ with
* $l = \cos \delta \sin \Delta \alpha$
* $m = \sin \delta\cos\delta_0 -\cos \delta\sin\delta_0\cos\Delta \alpha$
* $\Delta \alpha = \alpha - \alpha_0$

We can convert $lm$ to RA/DEC with
* $\delta = \sin^{-1}(m\cos\delta_0+\sqrt{1-l^2-m^2}\sin \delta_0)$
* $\alpha = \alpha_0 + \tan^{-1}\bigg( \frac{1}{\sqrt{1-l^2-m^2}\cos \delta_0 - m \sin \delta_0} \bigg)$

We have applied the above in the previous cell to obain our sky-model's  $l$ and $m$ coordinates.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.xlabel("$l$ [degrees]")
plt.ylabel("$m$ [degrees]")
plt.plot(l[0],m[0],"bx")
plt.hold("on")
plt.plot(l[1:]*(180/np.pi),m[1:]*(180/np.pi),"ro") 
counter = 1
for xy in zip(l[1:]*(180/np.pi)+0.25, m[1:]*(180/np.pi)+0.25):                                              
    ax.annotate(Flux_sources_labels[counter], xy=xy, textcoords='offset points',horizontalalignment='right',
                verticalalignment='bottom')  
    counter = counter + 1
        
plt.grid()
```

We now plot our sky model (it is now in $l$ and $m$ coordinates).

```python
point_sources = np.zeros((len(RA_sources)-1,3))
point_sources[:,0] = Flux_sources
point_sources[:,1] = l[1:]
point_sources[:,2] = m[1:]
```

We now create a convenient  numpy array in which to store our sky model

## Visibilites

We will now be creating a completely filled in $uv$-plane and sample it on the EW-baseline track we created in the first section. We will be ignoring the $w$-term for the sake of simplicity.

```python
u = np.linspace(-1*(np.amax(np.abs(u_60)))-10, np.amax(np.abs(u_60))+10, num=step_size, endpoint=True)
v = np.linspace(-1*(np.amax(abs(v_60)))-10, np.amax(abs(v_60))+10, num=step_size, endpoint=True)   
uu, vv = np.meshgrid(u, v)
zz = np.zeros(uu.shape).astype(complex)
```

We create the dimensions of our visibility plane.

```python
s = point_sources.shape
for counter in xrange(1, s[0]+1):
    A_i = point_sources[counter-1,0]
    l_i = point_sources[counter-1,1]
    m_i = point_sources[counter-1,2]
    zz += A_i*np.exp(-2*np.pi*1j*(uu*l_i+vv*m_i))
zz = zz[:,::-1]
```

We create our completely filled in visibitly plane. If we had a perfect interferometer we could sample the entire $uv$-plane, but due to the fact that we have only a finite amount of antennas this is not possible. Recall that our sky brightness $I(l,m)$ is related to to our visibilites $V(u,v)$ via the Fourier transform. For a bunch of point sources we can therefore write:

$V(u,v)=\mathcal{F}\{I(l,m)\} = \mathcal{F}\{\sum_k A_k \delta(l-l_k,m-m_k)\} = \sum_k A_k e^{-2\pi i (ul_i+vm_i)}$.

```python
u_track = u_60
v_track = v_60
z = np.zeros(u_track.shape).astype(complex)       

s = point_sources.shape
for counter in xrange(1, s[0]+1):
    A_i = point_sources[counter-1,0]
    l_i = point_sources[counter-1,1]
    m_i = point_sources[counter-1,2]
    z += A_i*np.exp(-1*2*np.pi*1j*(u_track*l_i+v_track*m_i))
```

Here we sample our visibility plane on the $uv$-track derived in the first section, i.e. $V(u_t,v_t)$.

```python
plt.imshow(zz.real,extent=[-1*(np.amax(np.abs(u_60)))-10, np.amax(np.abs(u_60))+10,-1*(np.amax(abs(v_60)))-10, np.amax(abs(v_60))+10])
plt.plot(u_60,v_60,"k")
plt.xlim([-1*(np.amax(np.abs(u_60)))-10, np.amax(np.abs(u_60))+10])
plt.ylim(-1*(np.amax(abs(v_60)))-10, np.amax(abs(v_60))+10)
plt.xlabel("u")
plt.ylabel("v")
plt.title("Real part of visibilities")
```

We now plot the real part of our visibility plane. On this plane we also draw our $uv$-track. As you can see we only sample a small portion of the visibility plane.

```python
plt.plot(z.real)
plt.xlabel("Timeslots")
plt.ylabel("Jy")
plt.title("Real: sampled visibilities")
```

We now plot the sampled visibilites as a function of time-slots, i.e $V(u_t(t_s),v_t(t_s))$.

```python
plt.imshow(zz.imag,extent=[-1*(np.amax(np.abs(u_60)))-10, np.amax(np.abs(u_60))+10,-1*(np.amax(abs(v_60)))-10, np.amax(abs(v_60))+10])
plt.plot(u_60,v_60,"k")
plt.xlim([-1*(np.amax(np.abs(u_60)))-10, np.amax(np.abs(u_60))+10])
plt.ylim(-1*(np.amax(abs(v_60)))-10, np.amax(abs(v_60))+10)
plt.xlabel("u")
plt.ylabel("v")
plt.title("Imaginary part of visibilities")
```
