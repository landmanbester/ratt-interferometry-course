Fourier transform of simple skies

```python
import numpy as np
import pylab as p
%matplotlib inline
```

Make a blank sky with a single, perfect delta function source near the centre of the image

```python
sky=np.zeros((128,128))
```

```python
sky[63,63]=1.
```

```python
fig = p.figure(figsize=(8,8))
p.imshow(sky,interpolation='nearest',cmap=p.cm.gray)
p.colorbar(shrink=0.5)
```

Perform 2 dimensional FFT of the sky. The shift is needed due to how the origin is defined. Our origin in the centre of the image.

```python
fig, axes = p.subplots(nrows=1, ncols=2, figsize=(16,8))
fig.tight_layout()
fourierSky=np.fft.fft2(np.fft.fftshift(sky))
p.subplot(1,2,1)
p.imshow(sky,interpolation='nearest',cmap=p.cm.gray)
p.colorbar(shrink=0.5)
p.subplot(1,2,2)
p.imshow(np.angle(fourierSky),cmap=p.cm.hsv)
p.colorbar(shrink=0.5)
#fig.savefig("point_src_0.png", bbox_inches='tight', pad_inches=0)
```

As we move the point away from the centre the phase rate increases out to the Nyquist sampling rate

```python
fig, axes = p.subplots(nrows=1, ncols=2, figsize=(16,8))
fig.tight_layout()
sky=np.zeros((128,128))
sky[48,48]=1.
fourierSky=np.fft.fft2(np.fft.fftshift(sky))
p.subplot(1,2,1)
p.imshow(sky,interpolation='nearest',cmap=p.cm.gray)
p.colorbar(shrink=0.5)
p.subplot(1,2,2)
p.imshow(np.angle(fourierSky),cmap=p.cm.hsv)
p.colorbar(shrink=0.5)
#fig.savefig("point_src_1.png", bbox_inches='tight', pad_inches=0)
```

Make a Gaussian source

```python
import scipy.signal
def gauss_kern(size):
    size = int(size)
    x, y = np.mgrid[-size:size+1, -size:size+1]
    g = np.exp(-(x**2/float(size)+y**2/float(size)))
    return g / g.sum()
sky=np.zeros((129,129))
sky[64,64]=1.
g = gauss_kern(128/2)
gaussSky=scipy.signal.convolve(sky,g, mode='full')
fourierSky=np.fft.fft2(np.fft.fftshift(gaussSky))
fig, axes = p.subplots(nrows=1, ncols=2, figsize=(16,8))
fig.tight_layout()
p.subplot(1,2,1)
p.imshow(gaussSky,interpolation='nearest',cmap=p.cm.gray)
p.colorbar(shrink=0.5)
p.subplot(1,2,2)
p.imshow(np.angle(fourierSky),cmap=p.cm.hsv)
p.colorbar(shrink=0.5)
#fig.savefig("gauss_srcs_0.png", bbox_inches='tight', pad_inches=0)
```

Create a ring source with multple Gaussians

```python
sky=np.zeros((129,129))
xx=np.array([32,29,22,12,0,-12,-22,-29,-32,-29,-22,-12,0,12,22,29])+64
yy=np.array([0,12,22,29,32,29,22,12,0,-12,-22,-29,-32,-29,-22,-12])+64
sky[xx,yy]=1.
g = gauss_kern(128/2)
gaussSky=scipy.signal.convolve(sky,g, mode='full')
fourierSky=np.fft.fft2(np.fft.fftshift(gaussSky))
fig, axes = p.subplots(nrows=1, ncols=2, figsize=(16,8))
fig.tight_layout()
p.subplot(1,2,1)
p.imshow(gaussSky,interpolation='nearest',cmap=p.cm.gray)
p.colorbar(shrink=0.5)
p.subplot(1,2,2)
p.imshow(np.angle(fourierSky),cmap=p.cm.hsv)
p.colorbar(shrink=0.5)
#fig.savefig("gauss_srcs_3.png", bbox_inches='tight', pad_inches=0)
```
