## 1st model fitting example

```python

# load in the data in file "data_eg2.txt" (2nd browser)


x_data = np.loadtxt('data_eg2.txt',usecols=[0]) # load 1st column
y_data = np.loadtxt('data_eg2.txt',usecols=[1]) # load 2nd column
yerr_data = np.ones(len(y_data)) * 0.1   #  uncertainty 0.1 for each data point

# let’s plot it to see what we’ve got:
pl.figure()
pl.plot(x_data,y_data,'o')
pl.errorbar(x_data,y_data,yerr=yerr_data,fmt="none",ecolor='r')
pl.xlabel('x_data')
pl.ylabel('y_data')
```

You now have a simulated dataset (**y_data**) which roughly equal to **sin(omega * x_data)**, where **omega** is the frequency of the sinusoid. Notice, that the data looks 'noisy', this is to simulate the effect of measurement error. Your job is to find **omega**. 

To do this we have to attempt a number of different values of **omega** and see which is the best fit to the data. We determine the best fit by calculating the sum of the square of the residuals of the data - model. 

So more formally, **residuals = sum[(data - model)^2 / error^2)] / (N_datapoints - N_parms -1) **

- where **N_datapoints** is the number of measurements we have; and 
- **N_parms** is the number of parameters that are free in the fitting process. More on this later and in formal texts. 

 so for our example: resid = np.sum((y_data - y_model)^2 / yerr_data^2)   # where "^" is "**" in python

 where **y_model** will be an array of trial "y" values that we generate by assuming a value of **omega**.

 So let's assume **omega** lies in the range 0.1 to 2. We'll step through "omega-space" in fairly coarse intervals of 0.3: 

```python
omega_max = 2
omega_min = 0.1
omega_step = 0.3
trialvalues = np.arange(omega_min,omega_max,omega_step)


for omega_i in trialvalues:

    y_model = np.sin(x_data*omega_i) # trial model with trial omega value
    resid = np.sum((y_data - y_model)**2/yerr_data**2)
    resid_r = resid/(len(y_model) - 1 - 1) # num of data points - num degrees of freedom (parameters to fit) - 1
    print 'for trial value of omega = %.1f:   \t  chi^2 = %.2f \t reduced chi^2 = %.1f'%(omega_i,resid,resid_r) 
    
    pl.figure(1)
    pl.plot(omega_i,resid_r,'ob')  # plot chi**2 residuals for each trial value of n
    pl.figure(2)
    pl.plot(x_data,np.sin(x_data*omega_i),'k',lw=0.9,alpha=0.7)

pl.plot(x_data,y_data,'o',ms=4,alpha=0.7)
pl.errorbar(x_data,y_data,yerr=yerr_data,fmt="none",ecolor='r')
pl.xlabel('x_data')
pl.ylabel('y_data')

# the output of which is:
```

two models are the 'best', but still don't really go through the data points within the uncertainties (which is an acid test for any model). We need to increase the resolution of our trial parameters. Next, we run precisely the same chunk of code as above, but this time we increase the number of trial frequency values by a factor of three:

```python
omega_max = 2
omega_min = 0.1
omega_step = 0.1
trialvalues = np.arange(omega_min,omega_max,omega_step)

for omega_i in trialvalues:

    y_model = np.sin(x_data*omega_i) # trial model with trial omega value
    resid = np.sum((y_data - y_model)**2/yerr_data**2)
    resid_r = resid/(len(y_model) - 2 - 1) # num of data points - num degrees of freedom (parameters to fit) - 1
    print 'for trial value of omega = %.1f:   \t  chi^2 = %.2f \t reduced chi^2 = %.1f'%(omega_i,resid,resid_r) 
    
    pl.figure(1)
    pl.plot(omega_i,resid_r,'ob')  # plot chi**2 residuals for each trial value of n
    pl.xlabel('trial omega value')
    pl.ylabel('reduced chi squared')
    pl.figure(2)
    pl.plot(x_data,np.sin(x_data*omega_i),'k',lw=0.6,alpha=0.5)

pl.plot(x_data,y_data,'o',ms=4,alpha=0.7)
pl.errorbar(x_data,y_data,yerr=yerr_data,fmt=None,ecolor='r')
pl.xlabel('x_data')
pl.ylabel('y_data')

# the output of which is:
```

Higher resolution gives better results (compare the chi square values with the coarser example. However, looping through parameters is not terribly efficient. You could skip the expense of for-loops and use the scipy levenburg-marquart solver, wrapped in a function called "curve_fit"

```python
from scipy.optimize import curve_fit

def func(x,omega):
    return np.sin(omega*x)

popt,pcov = curve_fit(func,x_data,y_data)
print 'LM solver finds omega = %.4f'%popt
print 'variance = %.6f'%pcov
print 'std dev  = %.4f \n \n \n '%pcov**0.5

print 'so the best-fit can be written as \n  omega = %.4f +- %.4f \n'%(popt,pcov**0.5)
# which give the following output: 
```

Compare this with the "grid-based" results you had previously.

## Fitting 2 parameters simultaneously

In the "grid-based" example you trialed values for omega only. Now we'll trial values of both omega and the amplitude of the sinusoid. The code is very simlar to previously, except now I have provided you with a simulated dataset which has unknown amplitude and frequency.

First, let's load and plot the simulated dataset as before:

```python
x_data = np.loadtxt('data_eg3.txt',usecols=[0]) # load 1st column
y_data = np.loadtxt('data_eg3.txt',usecols=[1]) # load 2nd column
yerr_data = np.ones(len(y_data)) * 1   #  uncertainty 0.1 for each data point

# let’s plot it to see what we’ve got:
pl.figure()
pl.plot(x_data,y_data,'o')
pl.errorbar(x_data,y_data,yerr=yerr_data,fmt="none",ecolor='r')
pl.xlabel('x_data')
pl.ylabel('y_data')
pl.savefig('plot5.png') 
```

 You'll need two for loops and to decide the range of trial values you wish to explore, based on the data above. 

```python
omega_max = 2
omega_min = 0.2
omega_step = 0.01 # note, much finer steps than previously
omega_trialvalues = np.arange(omega_min,omega_max,omega_step)

amp_max = 10
amp_min = 1
amp_step = 0.1 
amp_trialvalues = np.arange(amp_min,amp_max,amp_step)
                    

rchi = np.zeros([len(amp_trialvalues),len(omega_trialvalues)])# define an empty array, which we will store reduced-chi squared values in. 

for i in range(len(amp_trialvalues)):
  for j in range(len(omega_trialvalues)):

    y_model = amp_trialvalues[i] * np.sin(x_data*omega_trialvalues[j]) # trial model with trial omega value
    resid = np.sum((y_data - y_model)**2/yerr_data**2)
    resid_r = resid/(len(y_model) - 2 - 1) # num of data points - num degrees of freedom (parameters to fit) - 1
   
    rchi[i,j] = resid_r

```

The above code loops through trial values of both amplitude and frequency and stores the reduced chi-square values from each trial model. This produces an NxM array of reduced chi-square values, where N is the number of amplitude trials and M is the number of frequency trial values. Below we make an image to gauge which is the best pair of trial values. 

```python
# first we need to define the extent of the parameter space in each case.
# this takes the form of a list = [x.min(),x.max(),y.min(),y.max()]:
extent = [omega_trialvalues.min(),omega_trialvalues.max(),amp_trialvalues.min(),amp_trialvalues.max()]

# we now use the imshow command to plot an image (or 2D array)
pl.figure()
pl.imshow(np.log10(rchi),origin='lower',interpolation='nearest',extent=extent,aspect='auto')
# note I"ve taken the log of the rchi, which is purely for illustration purposes. The rest of the arguments are purely for presentation purposes. Experiment with them to see. 
cb = pl.colorbar()
cb.set_label('log(reduced chi^2)')
# but we want to find the acutal minimum, not just guess it by eye. To do this we use the numpy.where function:

chi_min = np.where(rchi == rchi.min())
print chi_min 
print 'note, the above are the indices of the trial value arrays, not the actual values of the best fit \n'
print 'best fit trial parameters are omega = %.2f; amplitude = %.2f \n \n'%(omega_trialvalues[chi_min[1]],amp_trialvalues[chi_min[0]])
                                                                                                                
pl.plot(omega_trialvalues[chi_min[1]],amp_trialvalues[chi_min[0]],'o',color='w')
pl.xlabel('trial omega values')
pl.ylabel('trial amplitude values')
pl.text(omega_trialvalues[chi_min[1]] + 0.05,amp_trialvalues[chi_min[0]],'min chi',color='grey') 
# above line adds some text to the plot at a specified coordinate
## NB: be aware of the python formalism for x,y order in images 
```

Given the best-fit parameters we have found, we now create what should be the best fit model and compare it with the data

```python

best_omega = omega_trialvalues[chi_min[1]]
best_amp = amp_trialvalues[chi_min[0]]
best_model = best_amp*np.sin(best_omega * x_data)

pl.figure()
pl.plot(x_data,y_data,'o',ms=4,alpha=0.7) # original data
pl.plot(x_data,best_model,label='best-fit model')
pl.errorbar(x_data,y_data,yerr=yerr_data,fmt=None,ecolor='r')
pl.xlabel('x_data')
pl.ylabel('y_data')
pl.legend()

```

Looks pretty good (the model passes within most of the errorbars). HOMEWORK: compare this with what curve_fit returns. 

## Estimating the Hubble parameter (expansion rate of the Universe)

Now let's fit something that is more physically interesting: like the expansion rate of the Universe. 

First, we need to import a function written by Ned Wright and James Schombert (if you don't know about CosmoCalc yet, visit this url immediately as you'll use this a great deal as an astronomer. http://www.astro.ucla.edu/~wright/CosmoCalc.html) 

```python
import nedwright   
```

This imports all functions within the file nedwright.py (note, it needs to be in the current directory or the PYTHONPATH)

You now have access to a function called cosmoCalc (try nedwright.cosmoCalc? )which takes four inputs:

- z = the redshift
- H0 = the value of the Hubble parameter at z=0
- omega_M = the energy-density of matter in the universe today (baryons + dark matter)
- omega_DE = the energy-density of dark energy in the Universe today   

so calling the following: 

```python
z = 0.05
H0 = 71
omega_M = 0.27
omega_DE = 0.73

kpc_DA = nedwright.cosmoCalc(z,H0,omega_M,omega_DE) 
```

you can figure out what these variables are from the code, however, the important one for this prac is the fourth: kpc_DA which gives the scale (kpc per arcsec) at redshift z for the given set of cosmological parameters: H0, omega_M, omega_DE

if I tell you that I have three measurements of the angular-spatial size scale at z = 0.05 and they are the following: 1.01 +- 0.05 kpc/arcsec; 0.99 +- 0.04; 0.98 +- 0.07. What is the best fit H0 value, assuming the same matter and dark energy densities from above? 

Hint: assume H0 is in the range 50-100 km/s/Mpc and take steps of 1 km/s/Mpc

NB: use the for loop (aka long) approach


```python
h0 = np.arange(50,100,5)
z = 0.05
omega_M = 0.27
omega_DE = 0.73  # omega's are set again just for clarity
data = np.array([1.01, 0.99, 0.98]) # the measurements of kpc per arcsec
data_err = np.array([0.05, 0.04, 0.07])
N_data_points = len(data) # I've given you three data points, which is the length of the data array
N_dof = 1     # number of degrees of freedom = 1, since we are only fitting for the Hubble Constant

chi = [] # empty list to collect chi values

for h in h0:
    print '\n \n \n_______________________________________________________'
    print ' \n Calculating values for current value of H0 = %.1f km/s/Mpc \n \n  '%h

    kpc_DA = nedwright.cosmoCalc(z,h,omega_M,omega_DE)
    chi.append( np.sum((data - kpc_DA)**2/data_err**2))


chi = np.array(chi) # note how you can use the value of lists and convert to arrays when/if needed
reduced_chi = chi/(N_data_points - N_dof -1)




pl.figure()
pl.plot(h0,reduced_chi,'o')
pl.xlabel('trial value of H$_0$')
pl.ylabel(r'$\chi^2_\nu$')
```

so where is chi^2 at it's minimum?  

```python
#print '\n \n \n \n ________________________________________'
print 'best fit is value of H0 where chi = chi.min()'
print 'this occurs at index %i of chi array '%np.where(chi == chi.min())
print 'so now find the best fit value of H0 by checking the value of h0[%i]'%np.where(chi == chi.min())
print 'so the best fit H0 for this Universe given the data is H0 = %.1f km/s/Mpc'%h0[np.where(chi==chi.min())]


# which gives the following output:
```
