# Averaging

Here we try to get a feel for the output of the correlator. Lets assume for the sake of simplicity that the output of the correlator is the averaging of a cosine plus a constant term, i.e $y = <C + \cos(2\pi f_c t)>$. Were $C$ represents the slowly varying term and $\cos(2\pi f_c t)$ the rapidly varying term. We will see that the longer we make our integration time the closer we get to $C$ our slowly varying term, which we assume here to be constant, i.e. $\lim_{t\rightarrow \infty} <C + \cos(2\pi f_c t)> = C$.

```python
f_c = 5 #frequency
t_0 = 0 #start time
t_1 = 0.15 #end time
C = 1 #constant
```

We set up our cosine equation parameters, the frequency is 5Hz, and we set $t_1$ to a value close to $t_0$.

```python
x = np.linspace(t_0,t_1,num=1000)
y = C+np.cos(2*np.pi*f_c*x)
plt.plot(x,y,"b")
y_m = np.mean(y)
plt.plot(x,np.ones(y.shape)*y_m,"r")
print y_m
```

We plot our cosine equation and the output of the correlator. The blue curve represents our cosine function and the red curve the 
output of the correlator (the average of the blue curve). Clearly we are still far from $C$, becasue $t_1$ is close to $t_0$. 

```python
f_c = 5
t_0 = 0
t_1 = 2.15
C = 1
```

We now increase the value of $t_1$.

```python
x = np.linspace(t_0,t_1,num=1000)
y = C+np.cos(2*np.pi*f_c*x)
plt.plot(x,y,"b")
y_m = np.mean(y)
plt.plot(x,np.ones(y.shape)*y_m,"r")
print y_m
```

We are now much closer to $C$.

```python
f_c = 5
t_0 = 0
t_1 = 20.15
C = 1
```

We now again increase the value of $t_1$.

```python
x = np.linspace(t_0,t_1,num=1000)
y = C+np.cos(2*np.pi*f_c*x)
plt.plot(x,y,"b")
y_m = np.mean(y)
plt.plot(x,np.ones(y.shape)*y_m,"r")
print y_m
```

We are now even closer to $C$. The output is approximately equal to $C$ if $(t_1-t_0)>>0$.

# Even Odd

Here we investigate why we need the sine correlator. Every function $f$ can be decomposed into the sum of an even $f_e$ and odd $f_o$ function. If we only apply the cosine correlator we will only pick up $f_e$, since the product of an even function
and an odd function is again odd, the symmetrical integral of an odd function is zero. We therefore use the sine correlator
to pick up the odd part of $f$. Here we have an example to illustrate this principal.

```python
x = np.linspace(-0.5,0.5,100)
y_even = x**2
y_odd = x**3
plt.plot(x,y_even,"b")
plt.plot(x,y_odd,"r")
plt.plot(x,y_even*y_odd,"g")
```

Here we have three functions, the blue graph represents an even function. The red graph represents an odd function. The green
graph is the product of the blue and red graphs, which is again an odd function. 

```python
y_even_int = np.trapz(y_even,x=x)
y_odd_int = np.trapz(y_odd,x=x)
y_prod_int = np.trapz(y_even*y_odd,x=x)
print y_even_int
print y_odd_int
print y_prod_int
```

The symmetrical integral of the green and red graphs are both zero.
