# Calibration

```python
import numpy as np
import pylab as plt
from scipy import optimize
%matplotlib inline
```

In this notebook we will present two simple approaches to perform direction independent calibration, namely 

* Levenberg-Marquardt - *Yatawatta, S., Kazemi, S., & Zaroubi, S. "GPU accelerated nonlinear optimization in radio interferometric calibration." In Innovative Parallel Computing (InPar), 2012 (pp. 1-6). IEEE.* 
* StEFCal - *Salvini, S., and Wijnholds, S. J. "Fast gain calibration in radio astronomy using alternating direction implicit methods: Analysis and applications." Astronomy & Astrophysics 571 (2014): A97*

Calibration is the procedure by which we estimate and remove the instrumental errors
induced onto interferometer data. Calibration is accomplished with:

$\textrm{min}_{\boldsymbol{G}}||\boldsymbol{R}- \boldsymbol{G}\boldsymbol{M}\boldsymbol{G}^H||$,

where

* $\boldsymbol{R}$ is the observed visibility matrix, entry $r_{pq}$ represents the visibility
observed by baseline $pq$.
* $\boldsymbol{M}$ is the model visibilites, generated by a calibration sky model.
* $\boldsymbol{G} = \textrm{diag}(\boldsymbol{g})$ is a diagonal matrix containing the
antenna gains on its diagonal.

In words the above equation expresses, find the antenna gains $\boldsymbol{g}$ that minimizes the difference between your observed and corrected visibilities.

We start by defining the class that we will use to accomplish this. We will discuss each function in detail later on in the notebook.

Note that we take into account the autocorrelations for simplicity of presentation, the code
below can easily be altered so that the autocorrelations are ignored. 

```python
class calibration():
   '''This class briefly shows you how to perform calibration using Levenberg-Marquardt and StEFCal.'''
   def __init__(self):
       pass  
   
   '''Unpolarized direction independent calibration entails finding the G that minimizes ||R-GMG^H||. 
   This function evaluates R-GMG^H.
   g is a vector containing the real and imaginary components of the antenna gains.
   r is a vector containing a vecotrized R (observed visibilities), real and imaginary.
   m is a vector containing a vecotrized M (predicted), real and imaginary.
   y is a vector containing the residuals.
   ''' 
   def err_func(self,g,r,m):
       Nm = len(r)/2
       N = len(g)/2
       G = np.diag(g[0:N]+1j*g[N:])
       R = np.reshape(r[0:Nm],(N,N))+np.reshape(r[Nm:],(N,N))*1j
       M = np.reshape(m[0:Nm],(N,N))+np.reshape(m[Nm:],(N,N))*1j
       T = np.dot(G,M)
       T = np.dot(T,G.conj())
       Y = R - T
       y_r = np.ravel(Y.real)
       y_i = np.ravel(Y.imag)
       y = np.hstack([y_r,y_i])
       return y

   '''This function finds argmin G ||R-GMG^H|| using Levenberg-Marquardt. It uses the optimize.leastsq scipy to perform
   the actual minimization.
   R is your observed visibilities matrx.
   M is your predicted visibilities.
   g the antenna gains.
   G = gg^H.''' 
   def create_G_LM(self,R,M):
       N = R.shape[0]
       temp =np.ones((R.shape[0],R.shape[1]) ,dtype=complex)
       G = np.zeros(R.shape,dtype=complex)
       g = np.zeros((R.shape[0],R.shape[2]),dtype=complex)
      
       for t in xrange(R.shape[2]):
           g_0 = np.ones((2*N,))
           g_0[N:] = 0
           r_r = np.ravel(R[:,:,t].real)
           r_i = np.ravel(R[:,:,t].imag)
           r = np.hstack([r_r,r_i])
           m_r = np.ravel(M[:,:,t].real)
           m_i = np.ravel(M[:,:,t].imag)
           m = np.hstack([m_r,m_i])
           g_lstsqr_temp = optimize.leastsq(self.err_func, g_0, args=(r, m))
           g_lstsqr = g_lstsqr_temp[0]          
           
           G_m = np.dot(np.diag(g_lstsqr[0:N]+1j*g_lstsqr[N:]),temp)
           G_m = np.dot(G_m,np.diag((g_lstsqr[0:N]+1j*g_lstsqr[N:]).conj()))           

           g[:,t] = g_lstsqr[0:N]+1j*g_lstsqr[N:]       
           G[:,:,t] = G_m
         
       return g,G

   '''This function finds argmin G ||R-GMG^H|| using StEFCal.
   R is your observed visibilities matrx.
   M is your predicted visibilities.
   imax maximum amount of iterations.
   tau stopping criteria.
   g the antenna gains.
   G = gg^H.'''
   def create_G_stef(self,R,M,imax,tau):
       N = R.shape[0]
       temp =np.ones((R.shape[0],R.shape[1]) ,dtype=complex)
       G = np.zeros(R.shape,dtype=complex)
       g = np.zeros((R.shape[0],R.shape[2]),dtype=complex)

       for t in xrange(R.shape[2]):
           g_temp = np.ones((N,),dtype=complex) 
           for i in xrange(imax):
               g_old = np.copy(g_temp)
               for p in xrange(N):
                   z = g_old*M[:,p,t]
                   g_temp[p] = np.sum(np.conj(R[:,p,t])*z)/(np.sum(np.conj(z)*z))
               if  (i%2 == 0):
                   if (np.sqrt(np.sum(np.absolute(g_temp-g_old)**2))/np.sqrt(np.sum(np.absolute(g_temp)**2)) <= tau):
                      break
                   else:
                      g_temp = (g_temp + g_old)/2
           
           G_m = np.dot(np.diag(g_temp),temp)
           G_m = np.dot(G_m,np.diag(g_temp.conj()))           

           g[:,t] = g_temp       
           G[:,:,t] = G_m
         
       return g,G   

   '''Creates the observed visibilities and the predicted visibilities
   point_sources - skymodel of point sources - (amplitude, l, m)
   u - the u coordinates of observation
   v - the v coordinates of observation
   error - the antenna gain error vector
   sig - the noise
   '''
   def create_R(self,point_sources,u,v,error,sig):
       R = np.zeros(u.shape)
       G = np.diag(error)
       for k in xrange(len(point_sources)):
           l_0 = point_sources[k,1]
           m_0 = point_sources[k,2]
           R = R + point_sources[k,0]*np.exp(-2*np.pi*1j*(u*l_0+v*m_0))
       for t in xrange(R.shape[2]):
           R[:,:,t] = np.dot(G,R[:,:,t])
           R[:,:,t] = np.dot(R[:,:,t],G.conj()) + sig*np.random.randn(u.shape[0],u.shape[1]) + sig*np.random.randn(u.shape[0],u.shape[1])*1j
       return R

   '''Creates very simplistic uv-tracks'''
   def simple_uv_tracks(self,Phi,num):
       lambda_v = 3e8/1.445e9
       length = 144/lambda_v
       l12 = length * Phi[0]
       l23 = length * Phi[1]
       l13 = length * Phi[2]
       t = np.linspace(-0.24,0.24,num=num)

       u = np.zeros((3,3,len(t)))
       v = np.copy(u)

       u[0,1,:] = l12*np.cos(2*np.pi*t)
       v[0,1,:] = l12*np.sin(2*np.pi*t)

       u[1,0,:] = -1*u[0,1,:]
       v[1,0,:] = -1*v[0,1,:]

       u[0,2,:] = l13*np.cos(2*np.pi*t)
       v[0,2,:] = l13*np.sin(2*np.pi*t)

       u[2,0,:] = -1*u[0,2,:]
       v[2,0,:] = -1*v[0,2,:]

       u[1,2,:] = l23*np.cos(2*np.pi*t)
       v[1,2,:] = l23*np.sin(2*np.pi*t)

       u[2,1,:] = -1*u[1,2,:]
       v[2,1,:] = -1*v[1,2,:]

       return u,v 
```

We start by creating a calibration object s and some $uv$-coverage. We do not explain how we 
create this $uv$-coverage as we assume that you are already familliar with it.

```python
s = calibration() #s for solver
u,v = s.simple_uv_tracks([2,3,5],1000)
   
# Plot uv coverage...
plt.plot(u[0,1,:],v[0,1,:],"b")
plt.hold("on")
plt.plot(u[1,0,:],v[1,0,:],"r")
plt.plot(u[0,2,:],v[0,2,:],"b")
plt.plot(u[2,0,:],v[2,0,:],"r")
plt.plot(u[1,2,:],v[1,2,:],"b")
plt.plot(u[2,1,:],v[2,1,:],"r")
plt.show()
```

We now create our observed $\boldsymbol{R}$ and model visibilities $\boldsymbol{M}$. We corrupt our model visibilites with a constant complex error (and add AWGN). The antenna error is equal to
$\mathbf{g}$.

```python
point_sources = np.array([(1,0,0),(0.5,(1*np.pi)/180,(0*np.pi)/180)])
error = np.array([1.2+1.3j,1.1-1.5j,-1.3+0.7j])
sig = 0.01
R = s.create_R(point_sources,u,v,error,sig)
M = s.create_R(point_sources,u,v,np.array([1,1,1]),0)
```

# Levenberg-Marquardt

We will be using least squares optimization to accomplish calibration. 

Suppose it is desired to fit a set of data $\left\{ \mathbf{m}_{i},\mathbf{r}_{i}\right\}$ to a known model, $\mathbf{r}=\mathbf{f}\left( \mathbf{m},\mathbf{g}^*\right)$ where $\mathbf{g}^*$ is a vector of parameters for the model that need to be found. A common method for determining which parameter vector gives the best fit to the data is to minimize the sum of squares of the residuals. The residual vector is denoted with $\mathbf{e}(\mathbf{g}^*,\mathbf{r},\mathbf{m}) = \mathbf{r} - \mathbf{f}\left( \mathbf{m},\mathbf{g}^*\right)$. The objective function to minimize is therefore: $\sum_i \mathbf{e}_i^2$. To use `optimize.leastsq` you need a function, `err_func`, that calculates the residual vector $\mathbf{e}$, you need an initual guess of the parameter vector $\mathbf{g}^*$, you need
the observed $\mathbf{r}$ and model input values $\mathbf{m}$.

For calibration the above variables become:

* $\mathbf{r} = [\textrm{vec}(\mathcal{R}\{\mathbf{R}\}),\textrm{vec}(\mathcal{I}\{\mathbf{R}\})]$
* $\mathbf{m} = [\textrm{vec}(\mathcal{R}\{\mathbf{M}\}),\textrm{vec}(\mathcal{I}\{\mathbf{M}\})]$
* $\mathbf{g}^* = [\mathcal{R}\{\mathbf{g}\},\mathcal{I}\{\mathbf{g}\}]$
* $\mathbf{f}\left( \mathbf{m},\mathbf{g}^*\right) = [\textrm{vec}(\mathcal{R}\{\mathbf{G}\mathbf{M}\mathbf{G}^H\}),\textrm{vec}(\mathcal{I}\{\mathbf{G}\mathbf{M}\mathbf{G}^H\})]$, where
$\mathbf{M} = \textrm{vec}^{-1}(\overline{\mathbf{m}})+i\textrm{vec}^{-1}(\underline{\mathbf{m}})$ and $\mathbf{G} = \textrm{diag}(\overline{\mathbf{g}}^*)+i\textrm{diag}(\underline{\mathbf{g}}^*)$. 

In the above bullets $\textrm{vec}(\cdot)$, $\textrm{vec}^{-1}(\cdot)$, $\overline{(\cdot)}$,
and $\underline{(\cdot)}$ respectively denote vectorization, matrization, the upper halve of
a vector and the lower halve of a vector. Moreover, $\mathcal{R}\{\cdot\}$ and $\mathcal{I}\{\cdot\}$ denote the real and imaginary part, respectively.

Below we calibrate via least squares (Levenberg-Marquardt). The input is $\mathbf{R}$ and $\mathbf{M}$, while the output is $\mathbf{g}$ and $\boldsymbol{\mathcal{G}}=\mathbf{g}\mathbf{g}^H$.

```python
glm,Glm = s.create_G_LM(R,M)
```

The above function works, by vectorizing the real and imaginary part of $\mathbf{R}$ and
storing the result in $\mathbf{r}$. The vector $\mathbf{m}$ is generated in a similar manner.
The error vector $\mathbf{e}$ is calculated by `s.err_func`. We initialize $\mathbf{g}^*$ with
$\mathbf{g}^*_0=[\mathbf{1},\mathbf{0}]$. We can then call 

`optimize.leastsq(self.err_func, g_0, args=(r, m))`.

The above function uses the *Levenberg-Marquardt* algorithm. We can now calculate $\mathbf{g} = \overline{\mathbf{g}}^*+i\underline{\mathbf{g}}^*$ and
$\boldsymbol{\mathcal{G}}=\mathbf{g}\mathbf{g}^H$. This is repeated for each observational time-slot.

```python
plt.plot(abs(glm[0]),"y")
plt.hold('on')
plt.plot(abs(glm[1]),"m")
plt.plot(abs(glm[2]),"c")
plt.show()
```

Above we have the absolute value of the gain solutions obtained via `leastsq`.

# StEFCal

StEFCal is an alternating direction implicit method. It works by first solving $\mathbf{G}^H$ holding $\mathbf{G}$ constant and then solving $\mathbf{G}$ holding $\mathbf{G}^H$ constant.
Since $\mathbf{R}- \mathbf{G}\mathbf{M}\mathbf{G}^H$ is Hermition the two steps are equivelant and
we therefore only need the following step:

$\mathbf{G}^{[i]} = \textrm{argmin}_{\mathbf{G}}||\mathbf{R}- \mathbf{G}^{[i-1]}\mathbf{M}\mathbf{G}^H||$.

We can now write

$||\mathbf{R} - \mathbf{Z}\mathbf{G}^H|| = \sqrt{\sum_{i}\mathbf{R}_{:,i}-\mathbf{Z}_{:,i}g_i^*}$, with $\mathbf{Z}^{[i]} = \mathbf{G}^{[i]}\mathbf{M}$. Here $()^{*}$ denotes conjugation. The above can easily be solved with

$g_p^{[i]} = \frac{\mathbf{R}^H_{:,p}\mathbf{Z}_{:,p}^{[i-1]}}{\left(\mathbf{Z}_{:,p}^{[i-1]}\right)^H\mathbf{Z}_{:,p}^{[i-1]}}$.

The function below performs the above iteratively. The input `imax` is the number of iterations,
and `tau` is the exit criteria.

```python
gs,Gs = s.create_G_stef(np.copy(R),M,20,1e-6)
```

Note how much faster StEFCal is in comparisson to LM.

```python
plt.plot(abs(gs[0]),"b")
plt.hold('on')
plt.plot(abs(gs[1]),"r")
plt.plot(abs(gs[2]),"g")
plt.show()
```

```python
np.absolute(error)
```

Note how similar the estimated absolute gains are to that obtained by LM.
