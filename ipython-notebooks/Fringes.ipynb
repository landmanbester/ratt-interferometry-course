# 1D Fringes and 1D Finite Bandwidth Attenuation

Our aim here is to investigate the 1D fringes that are being produced by our correlator and the attenuation that is caused by the fact that we have to observe within a finite bandwidth.

```python
import numpy as np
import pylab as plt
%matplotlib inline
```

# Initialization

```python
alpha = np.linspace(-90,90,1000)*(np.pi/180) #rad
f = 1.4*10**9 #observing frequency
c = 3*10**8 #speed of light
lam = c/f #wavelengt
b = 2+(1.0/7.0) #baseline length in m
x = b/lam #baseline length measured in wavelengths
alpha_c = 45*(np.pi/180) #rad
```

* $\alpha$ : The angle between zenith and the source direction
* $f$ : Observing frequency
* $c$ : The speed of light
* $\lambda$ : Observing wavelength
* $b$ : Baseline length
* $x$ : The baseline length measured in wavelenghts
* $\alpha_c$: Determines the position of the first null of the fringe-washing function

# Fringes

```python
y = np.cos(2*np.pi*x*np.sin(alpha))
```

The 1D fringe function is equal to $\cos(2\pi\frac{\boldsymbol{b}\cdot\boldsymbol{s}}{\lambda})=\cos(2\pi x\cos\theta)=\cos(2\pi x \sin\alpha)$.

```python
plt.plot(alpha*(180/np.pi),y)
plt.xlim([-90,90])
plt.xlabel(r"$\alpha$ [degrees]")
```

The 1D fringe function plotted in cartesian coordinates. Note that there are $x$ peaks between 0 and 90 degrees and that the graph is symmetrical.

```python
plt.axes([0, 0, 1, 1],polar=True)
plt.plot(alpha,np.abs(y))
plt.xlabel(r"$\alpha$ [derees]")
```

The 1D fringe function plotted in polar coordinates.

# Fringe Washing Function

The 1D fringe-washing function is equal to $F_w(\alpha)=\textrm{sinc}(\Delta f\tau_g)=\textrm{sinc}(\frac{\Delta f}{c}\boldsymbol{b}\cdot\boldsymbol{s})=\textrm{sinc}(\frac{\Delta f}{c}b\sin(\alpha))$. Note that it depends on the baseline length $b$ and not the baseline length measured in wavelengths $x$. So it is independent of your observing frequency and only depends on your channel bandwidth $\Delta f$.

```python
df = c/(b*np.sin(alpha_c))
print df
```

To determine the first null of $F_w(\alpha)$ we must remember that $\textrm{sinc}(x)=0$ when $x=1$, which implies that our
first null will be at $\alpha_c$ whenever $\Delta f = \frac{c}{b\sin(\alpha_c)}$. This supplies us a way of calculating the channel bandwidth 
that we require if we wish our first null to be at $\alpha_c$.

```python
yw = np.sinc((df/c)*b*np.sin(alpha))
```

The fringe washing function $F_w(\alpha)$.

```python
plt.plot(alpha*(180/np.pi),yw)
plt.xlim([-90,90])
plt.xlabel(r"$\alpha$ [degrees]")
```

The fringe washing function in Cartesian coordinates. Note that the first null appears at $\alpha_c$.

```python
plt.axes([0, 0, 1, 1],polar=True)
plt.plot(alpha,np.abs(yw))
plt.xlabel(r"$\alpha$ [derees]")
```

The fringe washing function in Polar coordinates. Note that the first null appears at $\alpha_c$.

```python
y2 = yw*y
```

The fringe washing function is applied to our fringe function, which leads to attenuation in the far-field.

```python
plt.plot(alpha*(180/np.pi),y)
plt.hold("on")
plt.plot(alpha*(180/np.pi),y2,"r")
plt.xlim([-90,90])
plt.xlabel(r"$\alpha$ [degrees]")
```

The red curves show our fringes after we have applied the fringe washing function (in Cartesion coordinates). See the attenuation in the
far field.

```python
plt.axes([0, 0, 1, 1],polar=True)
plt.plot(alpha,np.abs(y))
plt.hold("on")
plt.plot(alpha,np.abs(y2),"r")
plt.xlabel(r"$\alpha$ [derees]")
```

The red curves show our fringes after we have applied the fringe washing function (in polar coordinates). See the attenuation in the
far field.
